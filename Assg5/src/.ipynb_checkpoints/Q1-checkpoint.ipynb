```python
import numpy as np
from numpy import *
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
import random
from numpy.linalg import eig
import copy
from copy import deepcopy
import math
import sys
import operator 
eps = np.finfo(float).eps
```

```python
def train_test_split(df, test_size):
    if isinstance(test_size, float):
        test_size = round(test_size * len(df))
    indices = df.index.tolist()
    test_indices = random.sample(population=indices, k=test_size)
    
    test_df = df.loc[test_indices]
    train_df = df.drop(test_indices)
    return train_df, test_df
```

```python
def sigmoid(val):
    val = np.exp(-val+eps)
    return (1 / (1 + val))
```

```python
def initialization():
    weights1 = np.zeros((3, 784))
    weights2 = np.random.random((10,3))
    bias1 = [0 for i in range(3)]
    bias2 = [random.random() for i in range(10)]
    a1 = [0 for i in range(3)]
    h = [0 for i in range(3)]
    a2 = [0 for i in range(10)]
    y = [0 for i in range(10)]
    return weights1, weights2, bias1, bias2, a1, h, a2, y
```

```python
def hiddenLayer(trainData, weights1, a1, bias1, h):
    for i in range(3):
        for j in range(784):
            a1[i] += weights1[i][j] * trainData[j]
        a1[i] += bias1[i]
        h[i] = sigmoid(a1[i])
    return a1, h
```

```python
def outputLayer(trainData, weights2, h, a2, bias2):
    for i in range(10):
        for j in range(3):
            a2[i] += weights2[i][j] * h[j]
        a2[i] += bias2[i]
    return a2
```

```python
def softmax(a2, y):
    denom = 0
    for i in range(10):
        denom += np.exp(-a2[i]+eps)
    for i in range(10):
        y[i] = np.exp(-a2[i]+eps)/denom
    return y
```

```python
def update2(y, output, weights2, h, bias2):
    delJa2, eta = [], 0.2
    for i in range(10):
        if(i == output):
            delJa2.append(y[i]-1)
        else:
            delJa2.append(y[i])
    for i in range(10):
        for j in range(3):
            weights2[i][j] -= eta * delJa2[i] * h[j]
        bias2[i] -= eta * delJa2[i]
    return weights2, bias2, delJa2
```

```python
def update1(trainData, weights1, weights2, bias1, a1, delJa2):
    delJa1, eta = [], 0.2
    for i in range(3):
        for j in range(10):
            delJa1.append(delJa2[j] * weights2[j][i] * sigmoid(a1[i]) * (1 - sigmoid(a1[i])))
    for i in range(3):
        for j in range(784):
            weights1[i][j] -= eta * delJa1[i] * trainData[j]
        bias1[i] -= eta * delJa1[i]
    return weights1, bias1
```
