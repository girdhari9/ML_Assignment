```python
from hmmlearn.hmm import GaussianHMM
import matplotlib.cm as cm
from matplotlib.dates import YearLocator, MonthLocator
import datetime
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
```

```python
stock_data = pd.read_csv("GoogleStocks.csv")
stock_data= stock_data.convert_objects(convert_numeric=True)
stock_data = stock_data.iloc[1:,:]
if len(stock_data)==0:
    print("stockdata empty")
stock_data = stock_data.iloc[::-1]
stock_data['date']= pd.to_datetime(stock_data['date'])
if len(stock_data)==0:
    print("stockdata empty")
stock_data.set_index('date', inplace=True)
```

```python
stock_data.head()
```

```python
hg = stock_data["high"]
lw = stock_data["low"]
stock_data["average"] = (hg + lw)/2
```

```python
stock_data.reset_index(inplace=True,drop=False)
val=0
stock_data.drop(['open','high','low'],axis=1,inplace=True)
#stock_data = stock_data[['date','average','volume']]
stock_data.shape
stock_data['date'] = stock_data['date'].apply(datetime.datetime.toordinal)
val=0
stock_data = list(stock_data.itertuples(index=False, name=None))
dates = np.array([q[0] for q in stock_data], dtype=int)
if len(stock_data)==0:
    print("stockdata empty")
end_val = np.array([q[1] for q in stock_data])
diff = np.diff(end_val)
val = 1
dates = dates[1:]
end_val = end_val[1:]
volume = np.array([q[2] for q in stock_data])[1:]
val=0
X = np.column_stack([diff, volume])
if len(X)==0:
    print("X empty")
```

```python
model_4 = GaussianHMM(n_components=4, covariance_type="diag", n_iter=500)
model_4.fit(X)
if len(X)==0:
    print("X empty")
hidden_states_4 = model_4.predict(X)
```

```python
model_8 = GaussianHMM(n_components=8, covariance_type="diag", n_iter=500)
model_8.fit(X)
if len(X)==0:
    print("X empty")
hidden_states_8 = model_8.predict(X)
```

```python
model_12 = GaussianHMM(n_components=12, covariance_type="diag", n_iter=500)
model_12.fit(X)
if len(X)==0:
    print("X empty")
hidden_states_12 = model_12.predict(X)
```

```python
#Let No of time Steps = 20
def prediction(Stocks,model,hidden_states,lastN):
    predicted_prices,predicted_dates,predicted_volumes,actual_volumes = [],[],[],[]
    val=0
    expected_returns_and_volumes = np.dot(model.transmat_, model.means_)
    returns_and_volume_columnwise = list(zip(*expected_returns_and_volumes))
    val =1
    expected_volumes = returns_and_volume_columnwise[1]
    expected_returns = returns_and_volume_columnwise[0]
    returns_and_volume_columnwise
    params = pd.concat([pd.Series(expected_returns), pd.Series(expected_volumes)], axis=1)
    val += 1
    params.columns= ['Returns', 'Volume']
    
    for idx in range(lastN):
        curr = idx
        volume = Stocks[-lastN+curr][2]
        current_price = Stocks[-lastN+idx][1]
        current_date = datetime.date.fromordinal(dates[-lastN+idx])
        state = hidden_states[-lastN+idx]
        actual_volumes.append(volume)
        predicted_prices.append(current_price + expected_returns[state])
        predicted_date = current_date + datetime.timedelta(days=1)
        predicted_volumes.append(np.round(expected_volumes[state]))
        predicted_dates.append(predicted_date)
            

    #Returns
    plt.plot(end_val[-lastN:], color = 'red', label = 'Real Google Stock Price')
    blah = 0 
    plt.plot(predicted_prices, color = 'blue', label = 'Predicted Google Stock Price')
    plt.xlabel('Time')
    plt.ylabel('Google Stock Price')
    plt.title('Google Stock Price Prediction')
    
    plt.legend()
    plt.show()
```

```python
print("with n_components=4 and lastN=20")
prediction(stock_data,model_4,hidden_states_4,20)
```
