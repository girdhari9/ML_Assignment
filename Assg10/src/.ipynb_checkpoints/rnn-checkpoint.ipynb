```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import os
import random
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
def getPath():
    PATH = "../input_data"
    filename = os.listdir(PATH)
    filepath = os.getcwd()
    return (filepath + "/" + PATH + "/" + filename[0])
```

```python
PATH = getPath()
```

```python
def loadData(PATH):
    df = pd.read_csv(PATH)
    df = df.drop([0])
    return df
```

```python
df = loadData(PATH)
```

```python
def dataPreProcessing(df):
    cols = df.columns.tolist()
    newcols = [None for i in range(len(cols)+1)]
    if len(cols) != 6:
        return df
    else:
        reindex = [0, 3, 4, 5, 1, 1, 2]
        for i in range(len(cols) + 1):
            newcols[i] = cols[reindex[i]]
    return df[newcols]
```

```python
df = dataPreProcessing(df)
cols = df.columns
cols = cols.to_list()
```

```python
df.head()
```

```python
def train_test_split(df, test_size):
    train_indices = int(df.shape[0]*0.8)
    test_indices = int(df.shape[0]*0.8)-50
    
    train_df = np.array(df[ : train_indices])
    test_df = np.array(df[test_indices : ])
    
    train_df = pd.DataFrame(train_df, columns = cols)
    test_df = pd.DataFrame(test_df, columns = cols)
    return train_df, test_df
```

```python
traindf, testdf = train_test_split(df, 0.2)
```

```python
trainset = traindf.iloc[:,1:2].values
```

```python
sc = MinMaxScaler(feature_range = (0,1))
training_scaled = sc.fit_transform(trainset)
```

```python
len(training_scaled)
```

```python
x_train = []
y_train = []
```

```python
for i in range(60, len(training_scaled)):
    x_train.append(training_scaled[i-60:i, 0])
    y_train.append(training_scaled[i,0])
x_train,y_train = np.array(x_train),np.array(y_train)
```

```python
x_train.shape
```

```python
x_train = np.reshape(x_train, (x_train.shape[0], x_train.shape[1],1))
```

```python
def modelsCombo(units, epochs):
    combination = []
    for unit in units:
        for epoch in epochs:
            combination.append([unit, epoch])
    return combination
```

```python
params2 = modelsCombo([30,50,80], [20,50,75])
params3 = modelsCombo([30,50,80], [20,50,75])
```

```python
def prepareTestData(testdf):
    dataset_test = testdf
    real_stock_price = dataset_test.iloc[:,1:2].values
    dataset_total = pd.concat((df['open'],dataset_test['open']),axis = 0)
    inputs = dataset_total[len(dataset_total) - len(dataset_test)-60:].values
    inputs = inputs.reshape(-1,1)
    inputs = sc.transform(inputs)
    inputs.shape
    x_test = []
    for i in range(60,len(testdf)):
        x_test.append(inputs[i-60:i,0])
    x_test = np.array(x_test)
    x_test.shape
    x_test = np.reshape(x_test, (x_test.shape[0],x_test.shape[1],1))
    x_test.shape
    for i in range(len(real_stock_price)):
        real_stock_price[i] = float(real_stock_price[i][0])
    return x_test, real_stock_price
```

```python
x_test, real_stock_price = prepareTestData(testdf)
```

```python
def RNN(layers, unit, epoch):
    regressor = Sequential()
    regressor.add(LSTM(units = unit,return_sequences = True, input_shape = (x_train.shape[1],1)))
    regressor.add(Dropout(0.2))
    if layers == 3:
        regressor.add(LSTM(units = unit,return_sequences = True))
        regressor.add(Dropout(0.2))
    regressor.add(LSTM(units = unit))
    regressor.add(Dropout(0.2))
    regressor.add(Dense(units = 1))
    regressor.compile(optimizer = 'adam',loss = 'mean_squared_error')
    regressor.fit(x_train, y_train, epochs = epoch, batch_size = 32)
    return regressor
```

```python
from IPython.display import clear_output
predictPrices = []
for param in params2:
    clear_output()
    regressor = RNN(2, param[0], param[1])
    predictPrices.append(prediction(x_test, regressor))
```

```python
def prediction(x_test, regressor):
    predicted_price = regressor.predict(x_test)
    predicted_price = sc.inverse_transform(predicted_price)
    return predicted_price
```

```python
for predicted_price in predictPrices:
    showGraph(real_stock_price, predicted_price)
```

```python
def showGraph(real_stock_price, predicted_price):
    plt.plot(real_stock_price,color = 'red', label = 'Real Price')
    plt.plot(predicted_price, color = 'blue', label = 'Predicted Price')
    plt.title('Google Stock Price Prediction')
    plt.xlabel('Time')
    plt.ylabel('Google Stock Price')
    plt.legend()
    plt.show()
```

```python
predictPrices = []
for param in params3:
    clear_output()
    regressor = RNN(3, param[0], param[1])
    predictPrices.append(prediction(x_test, regressor))
```
