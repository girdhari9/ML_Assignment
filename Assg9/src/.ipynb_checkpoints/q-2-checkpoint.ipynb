# Part 1: Kernel Density Estimation

```python
from sklearn.neighbors import KernelDensity
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_digits
import pandas as pd
from sklearn.model_selection import GridSearchCV
import copy
from sklearn.datasets import load_digits
import math
from sklearn.mixture import GaussianMixture
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
def loadData():
    digits = load_digits()
    print(digits.data.shape)
    return digits
```

### For Reduced Dimensions = 25
#### Reduce 64-dimensional data to a lower dimension

```python
def PCAFunction(digits, comp):
    pca = PCA(n_components = comp, whiten = False)
    return pca.fit_transform(digits.data), pca
```

```python
def KDEImplement(data):
    grid = GridSearchCV(KernelDensity(), {'bandwidth': np.logspace(-1, 1, 20)}, cv=5)
    print(grid)
    grid.fit(data)
    return grid
```

```python
def showKDEvalue(grid):
    print("Best KDE value (Bandwidth) : {0}".format(grid.best_estimator_.bandwidth))
```

```python
def newData(pca, digits):
    KDE = grid.best_estimator_
    newPts = (pca.inverse_transform(KDE.sample(48, random_state=0))).reshape((4, 12, -1))
    real_data = digits.data[:48].reshape((4, 12, -1))
    return newPts, real_data
```

```python
if __name__ == "__main__":
    digits = loadData()
    data, pca = PCAFunction(digits, 25)
    grid = KDEImplement(data)
    showKDEvalue(grid)
    newPts, real_data = newData(pca, digits)
```

## Part 3: Inverse PCA transform

```python
def showDataPoints(real_data, newPts):
    fig, ax = plt.subplots(9, 12, subplot_kw=dict(xticks=[], yticks=[]))
    for j in range(0, 12, 1):
        ax[4, j].set_visible(False)
        for i in range(0, 4, 1):
            im = ax[i, j]
            reshapeData = real_data[i, j].reshape((8, 8))
            im = im.imshow(reshapeData, cmap = plt.cm.binary, interpolation='nearest')
                
            im.set_clim(0, 16)
            newReshape = newPts[i, j].reshape((8, 8))
            im = ax[i + 5, j]
            im = im.imshow(newReshape, cmap=plt.cm.binary, interpolation='nearest')
            im.set_clim(0, 16)

    ax[0, 5].set_title('Input data points', color="W")
    ax[5, 5].set_title('Digits after Inverse PCA using KDE Model', color = "W")
    plt.show()
```

```python
showDataPoints(real_data, newPts)
```

# Part 2: Gaussian Mixture Model based Density Estimation <br>
## Bayesian Information Criteria (BIC)

```python
digits = loadData()
data, pca = PCAFunction(digits, 15)
```

```python
def showPlot(data):
    n_components = np.arange(1, 51, 10)
    models, aics = [], []
    for n in n_components:
        models.append(GaussianMixture(n, covariance_type='full', random_state=0))
    for model in models:
        aics.append(model.fit(data).bic(data))
    plt.plot(n_components, aics)
```

```python
showPlot(data)
```

### For Reduced Dimensions = 11 (because, Least value from graph is 11)

```python
def gmmModel(data):
    gmm = GaussianMixture(11, covariance_type='full', random_state=0)
    gmm.fit(data)
    print(gmm.converged_)
    return gmm
```

```python
gmm = gmmModel(data)
```

```python
def dataPreProcess(digits, gmm, pca):
    new_data = pca.inverse_transform(gmm.sample(48)[0])
    new_data = new_data.reshape((4, 12, -1))
    real_data = digits.data[:48].reshape((4, 12, -1))
    return new_data, real_data
```

```python
new_data, real_data = dataPreProcess(digits, gmm, pca)
```

## Part 3: Inverse PCA transform

```python
showDataPoints(real_data, newPts)
```
